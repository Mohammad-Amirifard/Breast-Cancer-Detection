# Description of columns
* **site_id** - ID code for the source hospital.
* **patient_id** - ID code for the patient.
* **image_id** - ID code for the image.
* **laterality** - Whether the image is of the left or right breast.
* **view** - The orientation of the image. The default for a screening exam is to capture two views per breast.
* **age** - The patient's age in years.
* **implant** - Whether or not the patient had breast implants. Site 1 only provides breast implant information at the patient level, not at the breast level.
* **density** - A rating for how dense the breast tissue is, with A being the least dense and D being the most dense. Extremely dense tissue can make diagnosis more difficult. Only provided for train.
* **machine_id** - An ID code for the imaging device.
* **cancer** - Whether or not the breast was positive for malignant cancer. The target value. Only provided for train.
* **biopsy** - Whether or not a follow-up biopsy was performed on the breast. Only provided for train.
* **invasive** - If the breast is positive for cancer, whether or not the cancer proved to be invasive. Only provided for train.
* **BIRADS** - 0 if the breast required follow-up, 1 if the breast was rated as negative for cancer, and 2 if the breast was rated as normal. Only provided for train.
* **prediction_id** - The ID for the matching submission row. Multiple images will share the same prediction ID. Test only.
* **difficult_negative_case** - True if the case was unusually difficult. Only provided for train.

**In fact, for this amount of Nan/Null in age clomun, there is no need to fill them. Just we can delete the rows including these values. However, here, to learn how to deal with them, we used linear regreesion method to predict NaN/Null values in age column.**

```python
from sklearn.linear_model import LinearRegression
li = LinearRegression()
# With the help of following correlation matrix, we could understand that fetures like site_id, cancer, invasive, implant and difficult
# are more important than others on age target. So, we creat a dataframe with these feautre as input and age as target

new_dataframe_with_null = train_df[['site_id','cancer','invasive','implant','difficult_negative_case','age']]
new_dataframe_without_null = train_df[['site_id','cancer','invasive','implant','difficult_negative_case','age']].dropna()

# set x_train and y_train
x_train = new_dataframe_without_null.iloc[:,:5] # 'site_id','cancer','invasive','implant','difficult_negative_case'
y_train = new_dataframe_without_null.iloc[:,-1]  # 'age'

# Set x_test, inculdig parameters where age == null
x_test = new_dataframe_with_null[new_dataframe_with_null['age'].isnull()].drop(columns='age')

# Run model
li.fit(x_train,y_train)

# Predict on x_test
predicted = li.predict(x_test)
print('Here, you can see predicetd ages for missing values in main dataset')
print(f'-'*50)
print(predicted)
```

```python
# Setting variables for the model
main_directory = "/kaggle/working/Output/"
batch_size = 64
img_height = 256
img_width = 256
```

```python
# creating pipeline using "tf.keras.utils.image_dataset_from_directory"

class UseKerasUtils:
#      """This class has been provided to use tf.keras.utils.image_dataset_from_directory"
#         function to create a pipeline"""
    
    def __init__(self,main_directory, batch_size, img_height, img_width):
        self.data_dir = main_directory
        self.batch_size = batch_size
        self.img_height = img_height
        self.img_width = img_width

    def create_train_ds(self):
        train_ds = tf.keras.utils.image_dataset_from_directory(
        self.data_dir, 
        labels='inferred', # Since we have a main directories, including two subdirectories called ; Cacner and No_Cancer
        validation_split=0.2,  # keep 20 % of them to validate the result
        subset="training",
        seed=123,
        shuffle = True,
        color_mode="grayscale", # since our images are gray and have jsut one channel
        image_size=(self.img_height, self.img_width),
        batch_size=self.batch_size)
        return train_ds
    
    
    def create_val_ds(self):
        val_ds = tf.keras.utils.image_dataset_from_directory(
        self.data_dir, 
        labels='inferred',
        validation_split=0.2,  # keep 20 % of them to validate the result
        subset="validation",
        seed=123,
        shuffle = True,
        color_mode="grayscale",
        image_size=(self.img_height, self.img_width),
        batch_size=self.batch_size)
        return val_ds
```

```python
# Call functions and create two train and validation datasets 
first_pipeline = UseKerasUtils(main_directory, batch_size, img_height, img_width)
train_ds1 = first_pipeline.create_train_ds()
val_ds1 = first_pipeline.create_val_ds()
```

<p style="font-family: Times New Roman; font-size: 20px; font-style: bold; font-weight: bold; letter-spacing: 0px; color: #0000FF">5- Creating Model ResNet50</p>
<hr style="height: 1px; border: 1; background-color: #0000FF">

```python
 def identity_block(stage, block, inputs, filters, kernel_size = [(1,1),(3,3),(1,1)], strides=[(1,1),(1,1),(1,1)], padding=['valid','same','valid']):
        
        # define variables
        f1,f2,f3 = filters
        k1,k2,k3 = kernel_size
        s1,s2,s3 = strides
        p1,p2,p3 =padding
        conv_name_base =  'Conv_Stage' + str(stage) +"_Block_"+ block 
        bn_name_base = 'BatchNorm_Stage' + str(stage) +"_Block_"+ block 
        ac_name_base = 'Activation_Stage' + str(stage) + "_Block_" + block
    
        x_shortcut = inputs

        # First component of main path
        x = Conv2D(filters=f1, kernel_size=k1, strides=s1, padding=p1, name=conv_name_base + '_Component1')(inputs)
        x = BatchNormalization(name=bn_name_base + '_Component1')(x)
        x = Activation('relu',name=ac_name_base + '_Component1')(x)

        # Second component of main path
        x = Conv2D(filters=f2, kernel_size=k2, strides=s2, padding=p2, name=conv_name_base + '_Component2')(x)
        x = BatchNormalization(name=bn_name_base + '_Component2')(x)
        x = Activation('relu',name=ac_name_base + '_Component2')(x)

        # Third component of main path
        x = Conv2D(filters=f3, kernel_size=k3, strides=s3, padding=p3, name=conv_name_base + '_Component3')(x)
        x = BatchNormalization(name=bn_name_base + '_Component3')(x)

        # Final step: Add shortcut value to main path, and pass it through a RELU activation
        x = Add()([x, x_shortcut])
        x = Activation('relu',name=ac_name_base + '_Output')(x)
        return x
```

```python
 def convolutional_block(stage, block, inputs, filters,  kernel_size = [(1,1),(3,3),(1,1)], strides=[(2,2),(1,1),(1,1)], padding=['valid','same','valid']):
        
        # define variables

        f1,f2,f3 = filters
        k1,k2,k3 = kernel_size
        s1,s2,s3 = strides
        p1,p2,p3 =padding
        conv_name_base =  'Conv_Stage' + str(stage) +"_Block_"+ block 
        bn_name_base = 'BatchNorm_Stage' + str(stage) +"_Block_"+ block 
        ac_name_base = 'Activation_Stage' + str(stage) + "_Block_" + block
        x_shortcut = inputs
       
        # First component of main path 
        x = Conv2D(filters=f1, kernel_size=k1, strides=s1, padding=p1, name=conv_name_base + '_Component1')(inputs)
        x = BatchNormalization(name=bn_name_base + '_Component1')(x)
        x = Activation('relu',name=ac_name_base + '_Component1')(x)
       
        # Second component of main path (≈3 lines)
        x = Conv2D(filters=f2, kernel_size=k2, strides=s2, padding=p2, name=conv_name_base + '_Component2')(x)
        x = BatchNormalization(name=bn_name_base + '_Component2')(x)
        x = Activation('relu',name=ac_name_base + '_Component2')(x)

        # Third component of main path (≈2 lines)
        x = Conv2D(filters=f3, kernel_size=k3, strides=s3, padding=p3, name=conv_name_base + '_Component3')(x)
        x = BatchNormalization(name=bn_name_base + '_Component3')(x)
       
        ##### SHORTCUT PATH #### (≈2 lines)
        x_shortcut = Conv2D(filters=f3, kernel_size=k3, strides=s1, padding=p3, name=conv_name_base + '_Shortcut')(x_shortcut)
        x_shortcut = BatchNormalization(name=bn_name_base + '_Shortcout')(x_shortcut)
        
          # Final step: Add shortcut value to main path, and pass it through a RELU activation (≈2 lines)
        x = Add()([x, x_shortcut])
        x = Activation('relu',name=ac_name_base + '_Output')(x)
        return x
      
    
```

```python
def ResNet50(input_shape, outputClasses):
    
    # Define the input as a tensor with shape input_shape
    x_input = Input(input_shape,name= "My_Picture")
    
    x = Rescaling(scale=1./255)(x_input) # rescaling data to [0-1]
    # Zero-Padding
    x = ZeroPadding2D((3, 3), name = "ZeroPadding_Stage0")(x)

    # Stage 1
    x = Conv2D(64, kernel_size=(7, 7), strides=(2, 2), name="Conv_Stage1" )(x)
    x = BatchNormalization(name="BatchNorm_Stage1")(x)
    x = Activation('relu', name="Activation_Stage1")(x)
    x = ZeroPadding2D((1,1), name = "ZeroPadding_Stage1")(x)
    x = MaxPool2D((3, 3), strides=(2, 2), name= "MaxPooling_Stage1")(x)
  

    # Stage 2
    x = convolutional_block(stage=2, block='a',inputs=x, filters=[64, 64, 256], strides=[(1,1),(1,1),(1,1)])
    x = identity_block(stage=2, block='b', inputs=x, filters=[64, 64, 256])
    x = identity_block(stage=2, block='c',inputs=x, filters=[64, 64, 256])

    # Stage 3 
    x = convolutional_block(stage=3, block='a', inputs=x, filters=[128, 128, 512])
    x = identity_block(stage=3, block='b', inputs=x, filters=[128, 128, 512])
    x = identity_block(stage=3, block='c', inputs=x, filters=[128, 128, 512])
    x = identity_block(stage=3, block='d', inputs=x, filters=[128, 128, 512])

    # Stage 4
    x = convolutional_block(stage=4, block='a', inputs=x, filters=[256, 256, 1024])
    x = identity_block(stage=4, block='b', inputs=x, filters=[256, 256, 1024])
    x = identity_block(stage=4, block='c', inputs=x, filters=[256, 256, 1024])
    x = identity_block(stage=4, block='d', inputs=x, filters=[256, 256, 1024])
    x = identity_block(stage=4, block='e', inputs=x, filters=[256, 256, 1024])
    x = identity_block(stage=4, block='f', inputs=x, filters=[256, 256, 1024])

    # Stage 5
    x = convolutional_block(stage=5, block='a', inputs=x, filters=[512, 512, 2048])
    x = identity_block(stage=5, block='b', inputs=x, filters=[512, 512, 2048])
    x = identity_block(stage=5, block='c', inputs=x, filters=[512, 512, 2048])

    # AVGPOOL 
    x = AveragePooling2D(pool_size=(7, 7), padding='same', name = "AveragePooling")(x)

    # output layer
    x = Flatten(name= 'Faltten_OutputLayer')(x)
    x = Dense(outputClasses, activation='sigmoid', name='fc' + str(outputClasses))(x)

    # Create model
    model = Model(inputs=x_input, outputs=x, name='ResNet50')

     # Compile model
    my_optimizer = Adam(learning_rate=0.01)
    model.compile(loss='binary_crossentropy',optimizer=my_optimizer, metrics= 'accuracy')
    return model
```

```python

## Step 1: Define a function
def get_learning_rate_from_file(filename,epoch):
    with open(filename,'r') as f:
        for line in f.readlines():
            line = line.split('#',1)[0]
            if line:
                par =line.strip().split(':')
                j = int(par[0])
                lr = float(par[1])
                if j<= epoch and lr >0 :
                    learning_rate = lr
                else:
                    return learning_rate
                return learning_rate
        return learning_rate


## Step 2: Create a wrapper function to solve one input of the callback.LearninRateSchedule.
def get_schedule(filename):
    def scheduler(epoch):
        lr = get_learning_rate_from_file("/kaggle/working/learning_rate_file.txt",epoch)
        return lr
    return scheduler

## Step 3: call function
scheduler = get_schedule("/kaggle/working/learning_rate_file.txt")
My_LearningRateScheduler_callback = callbacks.LearningRateScheduler(scheduler)
```
